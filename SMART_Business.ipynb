```python
def highlight_priority(val):
    """
    Color High / Medium / Low priorities:
    - High   -> red
    - Medium -> yellow
    - Low    -> green
    """
    if val == "High":
        return "background-color: #ff4d4d; color: white;"   
    elif val == "Medium":
        return "background-color: #ffd633; color: black;"   
    elif val == "Low":
        return "background-color: #5cd65c; color: black;"   
    return ""


print("                                                      WATERFALL PLAN ")
display(df_waterfall.style.map(highlight_priority, subset=["Priority"]))

print("                                              AGILE SPRINT PLAN ")
display(df_agile.style.map(highlight_priority, subset=["Priority"]))

print("RACI MATRIX" " (R = Responsible , A = Accountable , C = Consulted , I = Informed )")
display(df_raci)
```

```python
#GANT DATA

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

gantt_data = [
    ("Initiation", 1, 4),
    ("Requirements Analysis", 4, 10),
    ("System Design", 10, 15),
    ("Sprint 1", 15, 20),
    ("Sprint 2", 20, 25),
    ("Sprint 3", 25, 30),
    ("Sprint 4", 30, 34),
    ("Sprint 5", 34, 38),
    ("Sprint 6", 38, 42),
    ("Sprint 7", 42, 46),
    ("Sprint 8", 46, 50),
]

df_gantt = pd.DataFrame(gantt_data, columns=["Task", "Start", "End"])
df_gantt["Duration"] = df_gantt["End"] - df_gantt["Start"]

#PLOT

plt.figure(figsize=(12, 6))

y_positions = np.arange(len(df_gantt))

for i, row in df_gantt.iterrows():
    plt.barh(
        y=i,
        width=row["Duration"],
        left=row["Start"],
        color=plt.cm.Set3(i / len(df_gantt)), 
        edgecolor="black"
    )

plt.yticks(y_positions, df_gantt["Task"], fontsize=10)
plt.xlabel("Project Day (150)", fontsize=12)
plt.title("TN CRM Project  Timeline (Waterfall + Agile Sprints)", fontsize=14)

#-----------------
plt.xticks(np.arange(0, 51, 5))

# -----------------
plt.grid(axis="x", linestyle="--", alpha=0.5)

plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```

```python
#VISUALIZATION CHARTS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# SPRINTS__________-----------------

# SPRINT RELATED TO WORKLOAD
sprint_labels = [f"Sprint {i}" for i in range(1, 9)]
sprint_effort = [12, 12, 12, 12, 12, 12, 14, 14]  

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(
    sprint_effort,
    labels=sprint_labels,
    autopct='%1.1f%%',
    startangle=90
)
ax.set_title("Sprint Effort Distribution")
ax.axis('equal') 
plt.show()


# Module PIE CHARTS ---------------------
module_labels = [
    "CRM UI & Customer Data",
    "Contract Workflow",
    "Delivery Module",
    "Invoicing",
    "Inventory",
    "Ticketing",
    "Power BI & Reporting"
]

module_complexity = [18, 15, 12, 15, 15, 10, 15] 
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(
    module_complexity,
    labels=module_labels,
    autopct='%1.1f%%',
    startangle=90
)
ax.set_title("Module Complexity / Effort Distribution")
ax.axis('equal')
plt.show()


#-------------------------RACI pie charts

# Departments
departments = [
    "PMO",
    "CRM Dev",
    "UI/UX",
    "Sales",
    "Legal",
    "Finance",
    "Logistics",
    "Warehouse",
    "Data Team",
    "IT / Integration",
    "Support",
    "Management"
]

# Ex
responsibility_count = [5, 6, 3, 4, 3, 4, 2, 2, 3, 4, 2, 5]

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(departments, responsibility_count)
ax.set_title("RACI Participation by Department")
ax.set_xlabel("Department")
ax.set_ylabel("Number of Key Responsibilities (R/A)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

#-----------------------
# Stakeholder lists
stakeholders = [
    ("Project Manager (IT)", "High", "High"),
    ("IT Department Staff", "High", "Medium"),
    ("External Consultants", "Medium", "Medium"),
    ("Sales Department", "High", "High"),
    ("Design Department", "Medium", "Medium"),
    ("Financial Department", "High", "Medium"),
    ("Company Management (Sponsors)", "High", "Low"),
]

influence_levels = ["Low", "Medium", "High"]
interest_levels = ["Low", "Medium", "High"]

# 3x3 matris 
heatmap_matrix = np.zeros((len(influence_levels), len(interest_levels)), dtype=int)

# Influence / Interest strings
influence_index = {level: i for i, level in enumerate(influence_levels)}
interest_index = {level: j for j, level in enumerate(interest_levels)}

for name, influence, interest in stakeholders:
    i = influence_index[influence]
    j = interest_index[interest]
    heatmap_matrix[i, j] += 1

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(heatmap_matrix)

# axises
ax.set_xticks(np.arange(len(interest_levels)))
ax.set_yticks(np.arange(len(influence_levels)))
ax.set_xticklabels(interest_levels)
ax.set_yticklabels(influence_levels)

ax.set_xlabel("Interest")
ax.set_ylabel("Influence")
ax.set_title("Stakeholder Risk Distribution (Influence vs Interest)")


for i in range(len(influence_levels)):
    for j in range(len(interest_levels)):
        value = heatmap_matrix[i, j]
        ax.text(j, i, str(value), ha="center", va="center")

fig.colorbar(im, ax=ax, shrink=0.8)
plt.tight_layout()
plt.show()
```
